## Part 1 - Web Scraping

       What we did in part 1 basically is using our Chorme broswer as a development tool, scraping the 4 GAFA companies' stock information from Yahoo Finance. First, we used requests to connect to the website and got data. Then, with the help of BeautifulSoup, we parsed the data we got and output it a decent dictionary form.

### Code

```python
#Code to scrape Yahoo Finance
import bs4
import requests
from bs4 import BeautifulSoup
def get_stock_info(x,link):
    r = requests.get(link)
    soup = bs4.BeautifulSoup(r.text,'html.parser')
    table = soup.find_all('table',{'data-test':"historical-prices"})
    # Use for loop and scrape the data
    for table in table:
        headers = []
        rows = table.find_all('tr')
        #Get headers
        for header in table.find('tr').find_all('th'):
            headers.append(header.text)
        #Get row information
        for row in table.find_all('tr')[1:]:
            values = []
            for col in row.find_all(['th','td']):
                values.append(col.text)
            if values:
                #Save data in dictionary
                Dict = {headers[i]: values[i] for i in range(len(values))} 
                x.append(Dict)
    #Print data
    for info in x:
        print(info)
```

## Part 2 - numpy + pandas

### Code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import pandas_datareader.data as web
```

```python
def build_dataframe(st):
    #get all the column names from the dictionary in part1 
    names = st[0].keys()
    #get all the stock information and put them in a list
    df = []
    for x in st[:-1]:
        df.append(list(x.values()))
    #change it to numpy form and then build a pandas data frame
    df = np.array(df)
    df = pd.DataFrame(df, columns = names)
    return df
```

### Daliy Percentage Change

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
#Add a new columns call Returns(**per dollor**)
amzn['Returns'] = (amzn['Adj Close']/amzn['Adj Close'].shift(1))-1
goog['Returns'] = (goog['Adj Close']/goog['Adj Close'].shift(1))-1
fb['Returns'] = (fb['Adj Close']/fb['Adj Close'].shift(1))-1
aapl['Returns'] = (aapl['Adj Close']/aapl['Adj Close'].shift(1))-1
Nasdaq['Returns'] = (Nasdaq['Adj Close']/Nasdaq['Adj Close'].shift(1))-1
```

```python
#Plot the histgram for Returns
amzn['Returns'].hist(bins = 60, label = 'Amazon' , alpha = 0.5, figsize = (14,7))
goog['Returns'].hist(bins = 60,label = 'Google', alpha = 0.5)
fb['Returns'].hist(bins = 60,label = 'Facebook', alpha = 0.5)
aapl['Returns'].hist(bins = 60,label = 'Apple', alpha = 0.3)
Nasdaq['Returns'].hist(bins = 60,label = 'Nasdaq', alpha = 0.3)
plt.legend()
```

### Cumulative Return

$$i_t = (1+r_t)i_{t-1}$$

```python
#Add a new columns call Returns(base is one dollar)
goog['Cumulative Return'] = (1 + goog['Returns']).cumprod()
amzn['Cumulative Return'] = (1 + amzn['Returns']).cumprod()
fb['Cumulative Return'] = (1 + fb['Returns']).cumprod()
aapl['Cumulative Return'] = (1 + aapl['Returns']).cumprod()
Nasdaq['Cumulative Return'] = (1 + Nasdaq['Returns']).cumprod()
```

```python
#Plot Cumulative Return (Amazon has the hishest cumulative return)
goog['Cumulative Return'].plot(label = 'Google',figsize = (15,11))
amzn['Cumulative Return'].plot(label = 'Amazon')
fb['Cumulative Return'].plot(label = 'Facebook')
aapl['Cumulative Return'].plot(label = 'Apple')
Nasdaq['Cumulative Return'].plot(label = 'Nasdaq')
plt.legend()
```
